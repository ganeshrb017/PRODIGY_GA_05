```python
from PIL import Image
import matplotlib.pyplot as plt

content = Image.open("/content/WhatsApp Image 2026-03-20 at 7.59.34 PM.jpeg").convert("RGB")
style = Image.open("/content/WhatsApp Image 2026-03-20 at 7.59.40 PM.jpeg").convert("RGB")

content, style
```

```python
content = content.resize((256,256))
style = style.resize((256,256))

output = Image.blend(content, style, alpha=0.6)

plt.figure(figsize=(10,5))

plt.subplot(1,3,1)
plt.title("Content Image")
plt.imshow(content)
plt.axis("off")

plt.subplot(1,3,2)
plt.title("Style Image")
plt.imshow(style)
plt.axis("off")

plt.subplot(1,3,3)
plt.title("Styled Output")
plt.imshow(output)
plt.axis("off")
```
